Para trabajar con objetos más complejos se deben definir una serie de funciones llamadas serializadores (en inglés: serializers). 

Por ejemplo, tenemos esta clase:

````python
class Mascota:
    def __init__(self, nombre, especie):
        self.nombre = nombre
        self.especie = especie
        
    def serialize(self):
        dict_object = {"nombre":self.nombre,
                       "especie":self.especie}
        return dict_object

class Alumno:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
        self.mascotas = []
        
    def serialize(self):
        mascotas = []
        for mascota in self.mascotas:
            mascotas.append(mascota.serialize())
        dict_object = {"nombre":self.nombre,
                       "edad":self.edad,
                       "mascotas": mascotas}
        
        return dict_object


alumn = Alumno("Pedro",20)
alumn.mascotas.append(Mascota("Tommy","gato"))
alumn.mascotas.append(Mascota("Caifás","perro"))
objeto_serializado = alumn.serialize()
objeto_serializado_cadena = json.dumps(objeto_serializado)
print(objeto_serializado)
````

Con el método loads podemos hacer el proceso inverso que se llama des-serializar.

````python
json.loads(objeto_serializado_cadena)
````

````python
def deserializar_alumno(cad_json):
    diccionario = cad_json #json.loads(cad_json)
    alumno = Alumno(nombre=diccionario["nombre"], edad=diccionario["edad"])
    for mascota in diccionario["mascotas"]:
        #mascota = json.loads(mascota)
        alumno.mascotas.append(Mascota(mascota["nombre"],mascota["especie"]))
    return alumno

alumno_des = deserializar_alumno(objeto_serializado)
alumno_des.mascotas
````
